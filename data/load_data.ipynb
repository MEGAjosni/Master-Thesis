### Load data

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
data_files = np.load('100134_1.8MW_55s_aaro_5_3.npz', allow_pickle=True)
data = {key: data_files[key] for key in data_files.files}
```

```python
print(data.keys())
```

```python
print(data['psin'].shape)
print(data['t'].shape)
print(data['ne'].shape)
```

```python
PSIN, T = np.meshgrid(data['psin'], data['t'], indexing='ij')

# Create the figure and axes
fig = plt.figure(figsize=(14, 6))

# Plot for Electron Density (ne)
ax1 = fig.add_subplot(121, projection='3d')
surf1 = ax1.plot_surface(PSIN, T, data['ne'], cmap='viridis', edgecolor='none')
cbar1 = fig.colorbar(surf1, ax=ax1, shrink=0.5, aspect=10)
cbar1.set_label('Electron Density (ne)', fontsize=12)
ax1.set_xlabel('Normalized Poloidal Flux (psin)', fontsize=12)
ax1.set_ylabel('Time (t)', fontsize=12)
ax1.set_zlabel('Electron Density (ne)', fontsize=12)
ax1.set_title('Electron Density Evolution', fontsize=14)

# Plot for Electron Temperature (te)
ax2 = fig.add_subplot(122, projection='3d')
surf2 = ax2.plot_surface(PSIN, T, data['te'], cmap='inferno', edgecolor='none')
cbar2 = fig.colorbar(surf2, ax=ax2, shrink=0.5, aspect=10)
cbar2.set_label('Electron Temperature (te)', fontsize=12)
ax2.set_xlabel('Normalized Poloidal Flux (psin)', fontsize=12)
ax2.set_ylabel('Time (t)', fontsize=12)
ax2.set_zlabel('Electron Temperature (te)', fontsize=12)
ax2.set_title('Electron Temperature Evolution', fontsize=14)

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```
