```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tanh = 'burger_upinn_tanh_clean'
sigmoid = 'burger_upinn_sigmoid_clean'
silu = 'burger_upinn_silu_clean'

tanh_data = np.load(f'logs/{tanh}.npy')
sigmoid_data = np.load(f'logs/{sigmoid}.npy')
silu_data = np.load(f'logs/{silu}.npy')
```

```python
# Extend data by last entry until it has langth 20000
while len(tanh_data) < 20000:
    tanh_data = np.concatenate((tanh_data, [tanh_data[-1]]))

while len(sigmoid_data) < 20000:
    sigmoid_data = np.concatenate((sigmoid_data, [sigmoid_data[-1]]))

while len(silu_data) < 20000:
    silu_data = np.concatenate((silu_data, [silu_data[-1]]))
```

```python
plt.figure( figsize=(10, 3) )
plt.plot(tanh_data[:, 0], label='Tanh', color='b')
plt.plot(sigmoid_data[:, 0], label='Sigmoid', color='r')
plt.plot(silu_data[:, 0], label='SiLU', color='g')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Total Loss')
# plt.title('Loss')
plt.yscale('log')
plt.ylim(5e-5, 1e-0)
plt.xlim(0, 20000)
plt.savefig('gfx/burger_act_loss.pdf', bbox_inches='tight', format='pdf')
plt.show()
```

```python
sa1 = 'burger_upinn_tanh_clean_softadapt_1'
sa05 = 'burger_upinn_tanh_clean_softadapt_05'
sa01 = 'burger_upinn_tanh_clean_softadapt_01'
sa005 = 'burger_upinn_tanh_clean_softadapt_005'
sa0 = 'burger_upinn_tanh_clean'
lam123 = 'burger_upinn_tanh_clean_123'
lam213 = 'burger_upinn_tanh_clean_213'

sa1_data = np.load(f'logs/{sa1}.npy')
sa05_data = np.load(f'logs/{sa05}.npy')
sa01_data = np.load(f'logs/{sa01}.npy')
sa005_data = np.load(f'logs/{sa005}.npy')
sa0_data = np.load(f'logs/{sa0}.npy')
lam123_data = np.load(f'logs/{lam123}.npy')
lam213_data = np.load(f'logs/{lam213}.npy')

while len(sa1_data) < 20000:
    sa1_data = np.concatenate((sa1_data, [sa1_data[-1]]))

while len(sa05_data) < 20000:
    sa05_data = np.concatenate((sa05_data, [sa05_data[-1]]))

while len(sa01_data) < 20000:
    sa01_data = np.concatenate((sa01_data, [sa01_data[-1]]))

while len(sa005_data) < 20000:
    sa005_data = np.concatenate((sa005_data, [sa005_data[-1]]))

while len(sa0_data) < 20000:
    sa0_data = np.concatenate((sa0_data, [sa0_data[-1]]))

while len(lam123_data) < 20000:
    lam123_data = np.concatenate((lam123_data, [lam123_data[-1]]))

while len(lam213_data) < 20000:
    lam213_data = np.concatenate((lam213_data, [lam213_data[-1]]))


plt.figure( figsize=(10, 3) )
plt.plot(sa1_data[:, 0], label=r'$\beta = 1$', color='b')
plt.plot(sa05_data[:, 0], label=r'$\beta = 0.5$', color='r')
plt.plot(sa01_data[:, 0], label=r'$\beta = 0.1$', color='g')
plt.plot(sa005_data[:, 0], label=r'$\beta = 0.05$', color='y')
plt.plot(lam123_data[:, 0], label=r'$\lambda = [0.2, 0.4, 0.8]$', color='c')
plt.plot(lam213_data[:, 0], label=r'$\lambda = [0.4, 0.2, 0.8]$', color='m')
plt.plot(sa0_data[:, 0], label=r'$\beta = 0$', color='k')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Total Loss')
# plt.title('Loss')
plt.yscale('log')
plt.ylim(5e-5, 1e-0)
plt.xlim(0, 20000)
plt.savefig('gfx/burger_sa_loss.pdf', bbox_inches='tight', format='pdf')
plt.show()
```
